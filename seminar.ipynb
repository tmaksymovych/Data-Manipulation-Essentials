# Датасет #2

Звернення служби "цілодобова варта". 
https://github.com/texty/datasets/tree/master/varta_for_publication

```python
%pip install geopandas
```

```python
import geopandas as gpd
import pandas as pd
import altair as alt
import ssl

ssl._create_default_https_context = ssl._create_unverified_context


gdf = gpd.read_file('https://polygons.openstreetmap.fr/get_geojson.py?id=361818&params=0')

gdf['geometry'] = gdf['geometry'].apply(lambda x: x.__geo_interface__)
df = pd.DataFrame(gdf)

df['coordinates'] = df['geometry'].apply(lambda x: x['coordinates'])
df['type'] = df['geometry'].apply(lambda x: x['type'])


alt.Chart(alt.Data(values=df.to_dict(orient='records'))).mark_geoshape().encode(
    shape='geometry:N'
).properties(
    width=600,
    height=400
).project('identity')

```
