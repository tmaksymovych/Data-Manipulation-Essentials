Створіть `Figure` і додайте туди 4 панелі (2х2) за допомогою методу `subplots`. 

Метод subplots поверне вам **масив** з обʼєктів, кожен з яких репрезентує певну панель, і з кожним з цих обʼєктів можна працювати окремо - наприклад, додавати графіки, змінювати заголовки і т.д.

Помістіть у панелі графіки (по одному) різних функцій - $y = x$, $y=\frac{x}{2}, $y = \frac{x^2}{10}$, $y= \sin{x}$. 

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.flatten()
axs[0].plot(x, x)
axs[1].plot(x, x/2)
axs[2].plot(x, x*x/10)
axs[3].plot(x, np.sin(x))
```

Можливо, ви помітили, що "щось не так з масштабом". Дійсно, панелі, які ми створили, не ділять між собою шкали - у кожного з них масштаб геть відрізняється від іншого. 

Щоб "синхронізувати" їх, вам знадобиться параметр `sharey=True` (і звісно, якщо у вас відрізняються границі x - то і `sharex=True`). 

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex=True, sharey=True)
axs = axs.flatten()
axs[0].plot(x, x)
axs[0].set_title("y=x")
axs[1].plot(x, x/2)
axs[2].plot(x, x*x/10)
axs[3].plot(x, np.sin(x))
plt.tight_layout()
```

Нарешті, деякі операції хочеться застосувати до всіх підграфіків. Наприклад, як зробити, щоб сітка показувалась для всіх панелей? Звісно що використати цикл!

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex=True, sharey=True)
axs = axs.flatten()

axs[0].plot(x, x)
axs[0].set_title("y=x")

axs[1].plot(x, x/2)
axs[1].set_title("y=x/2")

axs[2].plot(x, x*x/10)
axs[2].set_title("y=x^2/10")

axs[3].plot(x, np.sin(x))
axs[3].set_title("y=sin(x)")

plt.tight_layout()
for i in range(4):
    axs[i].grid(linestyle="dotted", color="#bfc8d0")
    axs[i].axhline(0, color="darkgrey", linewidth=1)
```

Намалюйте у комірках графіки довільних функцій. І заодно спробуйте наступне: що, якщо використати функцію, яка не означена у деяких місцях (наприклад, $y = \sqrt{x}$ або $y = x^x$)?

Крім того, подивіться, як працюють для такого складного випадку параметри `sharex=True, sharey=True`

```python
layout = [["A", "B"],
        ["A", "C"],
        ["D", "D"]]



fig, axd = plt.subplot_mosaic(layout, figsize=(10, 10))

axd['A'].plot(x, np.sqrt(np.maximum(x, 0)), "b-", label='y = √x')
axd['A'].set_title('Square root function')
axd['A'].legend()

axd['B'].plot(x, x**np.where(x > 0, x, np.nan), 'r-', label='y = x^x')
axd['B'].set_title('x to the power of x')
axd['B'].legend()

axd['C'].plot(x, x, 'g-', label='y = x')
axd['C'].set_title('y=x')
axd['C'].legend()

axd['D'].plot(x, np.log(np.abs(np.where(x != 0, x, np.nan))), 'm-', label='y = ln|x|')
axd['D'].set_title('Natural logarithm')
axd['D'].legend()

for ax in axd.values():
    ax.grid(True)
    ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax.axvline(x=0, color='k', linestyle='-', alpha=0.3)
```

Спробуйте створити набір графіків, який би зображував будь-що, для чого була б корисна візуалізація. Це може бути фізичний закон, красивий математичний факт, економічні принципи - загалом будь-що, що вам видається вартим візуалізації. 

Головні вимоги:
- Графіків має бути декілька (мінімум 2) і вони мають бути поєднані у спільну Figure;
- Кожен графік має містити суміщати кілька ліній (а не просто умовний $y=x^2$);
- Кожен графік має бути підписаним, з додатковими позначками там, де це має сенс. В ідеалі, графік мав би "читатися" без додаткових пояснень, а підписи/позначки мали б підкреслювати основний факт з відповідного графіку;
- Стилі графіків мають бути відмінні від тих, що за змовчуванням. Підберіть кольори, які вдало підходять до діаграми і які на ваш погляд є влучними.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h_km = np.linspace(0, 20, 200)
temp = 15
temp_gradient = 6.5# °с/км(темп спаду температури в тропосфері)
tropopause = 11# км(шар атмосфери, де температура перестає знижуватися)

temp_c = np.where(h_km <= tropopause,
                  temp - temp_gradient * h_km,
                  temp - temp_gradient * tropopause)

pressure = 1013#гПа  
hieght = 7.5

pressure_hpa = pressure * np.exp(-h_km / hieght)
```

```python
layout = [["A"],
          ["B"]]
fig, axd = plt.subplot_mosaic(layout, figsize=(10,10), gridspec_kw={'hspace': 0.3} )

line_temp, = axd["A"].plot(h_km, temp_c, color='darkgreen', linestyle='-', label="Temperature")
axd["A"].set_ylabel('Tempereture (°C)',fontsize=15)
axd["A"].set_xlabel("Height (km)", fontsize=15)
axd["A"].set_title("Temperature change with altitude", fontsize=25)
axd["A"].axvline(tropopause, color='gray', linestyle="--")
axd["A"].text(5, -5, "Troposphere", color="black", rotation=40, fontsize=15)
axd["A"].text(14, -45, "Tropopause", color="black", fontsize=15)
axd["A"].invert_yaxis()
axd["A"].legend(fontsize=14)

axd["B"].plot(h_km, pressure_hpa, color="darkred")
axd["B"].set_title("Pressure change with altitude", fontsize=25)
axd["B"].set_ylabel('Pressure (Hpa)',fontsize=15)
axd["B"].set_xlabel("Height (km)", fontsize=15)
axd["B"].invert_yaxis()
axd["B"].legend()


for ax in axd.values():
    ax.grid(linestyle="dotted", color="#020202")
    ax.axhline(0, color="white", linewidth=1)

```
